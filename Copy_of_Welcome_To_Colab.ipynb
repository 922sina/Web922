```

import pandas as pd
import matplotlib.pyplot as plt

```

```
school =pd.read_csv("/content/School_Attendance_by_Student_Group_and_District__2021-2022.csv")
```

```
categories_count =school['Category'].value_counts()
plt.figure(figsize=(14,6))
plt.bar(categories_count.index,categories_count)
plt.title('Count plot of Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
